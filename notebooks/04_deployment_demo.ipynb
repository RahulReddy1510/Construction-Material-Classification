# 04. Deployment Demo

Final end-to-end demo of the Construction Material Classifier running in a simulated edge environment.

```python
import sys
sys.path.append('..')
from deployment.inference import InferenceRunner
import matplotlib.pyplot as plt
import cv2

MODEL_PATH = '../checkpoints/efficientnet_best.pth'
IMAGE_PATH = '../data/synthetic/test/concrete/concrete_0000.jpg'
class_names = ['concrete', 'brick', 'metal', 'wood', 'stone']

runner = InferenceRunner(MODEL_PATH, class_names)
label, score = runner.predict(IMAGE_PATH)

img = cv2.imread(IMAGE_PATH)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
plt.title(f"Prediction: {label} ({score*100:.1f}%)")
plt.axis('off')
plt.show()
```
