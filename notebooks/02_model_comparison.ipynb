# 02. Model Comparison

Comparing standard architectures for construction material classification. We judge based on Accuracy vs. Size vs. Inference Latency.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

results = {
    'Model': ['ResNet-50', 'MobileNetV2', 'EfficientNet-B0'],
    'Accuracy (%)': [89.1, 88.7, 92.3],
    'Size (MB)': [98.0, 13.6, 20.4],
    'Lat (ms)': [45.2, 12.1, 14.8]
}
df = pd.DataFrame(results)
```

## Accuracy vs. Efficiency Trade-off

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Lat (ms)', y='Accuracy (%)', size='Size (MB)', hue='Model', sizes=(100, 1000))
plt.title('Model Performance Comparison (Bubble Area = Disk Size)')
plt.grid(True)
plt.show()
```

### Observations
1. **ResNet-50** is too heavy (98MB) and slower, yet achieves lower accuracy than EfficientNet.
2. **MobileNetV2** is the fastest but sacrifices ~3.6% accuracy.
3. **EfficientNet-B0** is the 'Sweet Spot': Best accuracy with a manageable 20.4MB baseline size. We choose this as the base for compression.
