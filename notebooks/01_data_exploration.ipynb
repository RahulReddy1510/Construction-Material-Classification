# 01. Data Exploration

This notebook explores the dataset used for the Construction Material Classifier. We analyze both the synthetic fallback data and the characteristics of real material textures.

```python
import os
import cv2
import matplotlib.pyplot as plt
import numpy as np
from glob import glob

DATA_DIR = '../data/synthetic/'
classes = ['concrete', 'brick', 'metal', 'wood', 'stone']
```

## Class Distribution

```python
counts = {cls: len(glob(os.path.join(DATA_DIR, 'train', cls, '*'))) for cls in classes}
plt.figure(figsize=(10, 5))
plt.bar(counts.keys(), counts.values(), color='skyblue')
plt.title('Training Samples per Class')
plt.show()
```

## Visualizing Material Textures

```python
fig, axes = plt.subplots(1, 5, figsize=(20, 5))
for i, cls in enumerate(classes):
    img_path = glob(os.path.join(DATA_DIR, 'train', cls, '*.jpg'))[0]
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    axes[i].imshow(img)
    axes[i].set_title(cls)
    axes[i].axis('off')
plt.show()
```

### Texture Analysis
- **Concrete**: High-frequency noise with characteristic horizontal streaks.
- **Brick**: Structural grid with mortar transitions.
- **Metal**: Anisotropic grain and simulated specular highlights.
- **Wood**: Sine-wave grain patterns with procedural knots.
- **Stone**: Voronoi-based polygonal structures representing aggregate.
